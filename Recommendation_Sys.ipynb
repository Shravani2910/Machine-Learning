```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("interview_movies_ratings.csv")
df.head()
```

```python
# Creating user item matrix

user_matrix = df.pivot_table(
    index="user_id",
    columns="movie_id",
    values="rating"
)
user_matrix = user_matrix.fillna(0)
```

```python
user_matrix.head()
```

```python
# Collaborative Filtering

from sklearn.metrics.pairwise import cosine_similarity
item_similarity = cosine_similarity(user_matrix.T)
item_sim = pd.DataFrame(
    item_similarity,
    index=user_matrix.columns,
    columns=user_matrix.columns
)
```

```python
# Item based movie recommendation

def recommend_item_based(movie_id,top_n=5):
    similar_movies =(
        item_sim[movie_id]
        .sort_values(ascending=False)[1:top_n+1])
    return similar_movies
```

```python
# user based movie recommendation

user_similarity = cosine_similarity(user_matrix)
user_Sim = pd.DataFrame(
    user_similarity,
    index=user_matrix.index,
    columns=user_matrix.index
)
```

```python
# Matrix Factorization (SVD)

from sklearn.decomposition import TruncatedSVD
svd = TruncatedSVD(n_components=5,random_state=42)
latent_matrix = svd.fit_transform(user_matrix)
pred_ratings = svd.inverse_transform (latent_matrix)
pred_df = pd.DataFrame(
    pred_ratings,
    index=user_matrix.index,
    columns=user_matrix.columns
)
```

```python
# Generating Top-N recommedations using SVD

def recommend_for_user(user_id, n=10):
    user_predictions = pred_df.loc[user_id]
    watched = user_matrix.loc[user_id]

    recommended = user_predictions[watched == 0]

    return recommended.sort_values(ascending=False).head(n)
```

```python
# RMSE Evaluation (Root Mean Square)

from sklearn.metrics import mean_squared_error
import numpy as np

true_values = user_matrix.values
pred_values = pred_df.values

rmse = np.sqrt(mean_squared_error(true_values, pred_values))
rmse
```
